```python
import numpy as np
import matplotlib.pyplot as plt

data_house_kobuki_05 = np.loadtxt('../../dataset/data_house_kobuki_05.csv', delimiter=',')
data_house_kobuki = np.loadtxt('../../dataset/data_house_kobuki.csv', delimiter=',')
data_marsyard2020 = np.loadtxt('../../dataset/data_marsyard2020.csv', delimiter=',')
data_marsyard2020_1 = np.loadtxt('../../dataset/data_marsyard2020_1.csv', delimiter=',')

s = 1
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

for p in data_marsyard2020:
    p1, p2, p3, trav = p
    c = 'red'
    if trav > 0:
        c = 'green'
    ax.scatter(p1, p2, 1/p3, s=s, color=c)

ax.set_xlabel('Inclination')
ax.set_ylabel('Variation')
ax.set_zlabel('Number of points')

plt.show()
```

```python
def f_inc(inc):
    return np.sqrt(1 / (1 + inc**2))

for p in data_house_kobuki_05:
    inc, sigma, n, trav = p
    y = sigma
    x = f_inc(inc)
    c = 'red'
    if trav > 0:
        c = 'green'
    if sigma > 0.005:
        continue
    plt.scatter(x, y, s=s, color=c)

plt.title('Kobuki Bot in House. Cell size 0.5m')
plt.xlabel('Inclination cosine')
plt.ylabel('Variation')
plt.show()
```
