```python
import numpy as np
import matplotlib.pyplot as plt

s = 1
data_house = np.loadtxt('../../dataset_realtime/house_1.csv', delimiter=',')

def f_inc(inc):
    return np.sqrt(1 / (1 + inc**2))

for p in data_house:
    inc, sigma, n, z_range, trav = p
    y = sigma
    x = f_inc(inc)
    c = 'red'
    if trav > 0:
        c = 'green'
    if sigma > 0.005:
        continue
    plt.scatter(x, y, s=s, color=c)

plt.title('Kobuki Bot in House. Cell size 0.5m')
plt.xlabel('Inclination cosine')
plt.ylabel('Variation')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

segment_0 = np.loadtxt('../../clouds_realtime/segment_0.csv', delimiter=',')

s = 1
fig = plt.figure()
ax4 = fig.add_subplot(projection='3d')

for i in range(0, segment_0.shape[0], 100):
    x, y, z = segment_0[i, :]
    c = 'blue'
    ax4.scatter(x, y, z, s=s, color=c)

ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.set_zlabel('z')

plt.show()
```
